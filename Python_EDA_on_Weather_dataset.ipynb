```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as p
import seaborn as sns
```

```python
season=pd.read_csv("weather.csv")
```

```python
# VISUALISATION
# create a group rain today status according to humidity, pressure and temp
group_rain_today=season.groupby("RainToday").agg({"Humidity9am" : "mean",
    "Humidity3pm" : "mean", "Pressure9am" : "mean",
    "Pressure3pm" : "mean", "Temp9am" : "mean", "Temp3pm" : "mean"})
```

```python
group_rain_today
```

```python
p.figure(figsize=(4,4))
sns.heatmap(group_rain_today, annot=True)
p.title("Today Rain Status")
p.show()
```

```python
# create a rain tomorrow status according to humidity and temperature
group_rain_tomorrow=season.groupby("RainTomorrow").agg({"Humidity9am" : "mean",
    "Humidity3pm" : "mean", "Temp9am" : "mean", "Temp3pm" : "mean"})
```

```python
group_rain_tomorrow
```

```python
p.figure(figsize=(5,5))
sns.heatmap(group_rain_tomorrow, annot=True)
p.title("Tomorrow Rain Status")
p.show()
```

```python
# visulaizatino of the countplot on entrie dataframe of weather 
p.figure(figsize=(7,5))
datalist=[season["WindSpeed9am"], season["WindSpeed3pm"], season["Humidity9am"], 
season["Humidity3pm"], season["Temp9am"], season["Temp3pm"]]
ax=sns.countplot(data=datalist)
ax.bar_label(ax.containers[0])
p.show()
```
